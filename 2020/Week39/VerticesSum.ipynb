```julia
using LazySets, Plots
```

```julia
Z1 = convert(Zonotope, BallInf(zeros(2), 1.0))
Z2 = copy(Z1)
```

```julia
plot(Z1, lab="Z1")
plot!(Z2, lab="Z2")
```

```julia
V = VPolygon(vertices_list(Z1) .+ vertices_list(Z2))
plot!(V, lab="V")
```

```julia
Y = Zonotope(zeros(2), Z1.generators .+ Z2.generators)
```

```julia
plot!(Y, lab="Y")
```

```julia
Z1 = convert(Zonotope, [1 2; 3 2] * BallInf(zeros(2), 1.0))
Z2 = copy(Z1)

V = VPolygon(vertices_list(Z1) .+ vertices_list(Z2))

Y = Zonotope(zeros(2), Z1.generators .+ Z2.generators)

plot(Z1, lab="Z1")
plot!(Z2, lab="Z2")
plot!(V, lab="V")
plot!(Y, lab="Y")
```

```julia
(V  Y) && (Y  V)
```
