## Concave hull

```julia
using ConcaveHull
using Plots
```

```julia
y(x) = cos(x)
xdom = range(0, 2.5pi, length=80)
yval = [cos(x) + 0.2*rand() for x in xdom];

plot(xdom, yval, seriestype=:path, marker=:dot, lab="Data", legend=:bottomright)

points = [[xi, yi] for (xi, yi) in zip(xdom, yval)];
conchull = concave_hull(points)

plot!(conchull, c=:red, lab="Concave hull", lw=1.5)
```

## Convex hull

```julia
using LazySets
```

```julia
plot(xdom, yval, seriestype=:path, marker=:dot, lab="Data", legend=:bottomright)

convhull = convex_hull(points)
phull = VPolygon(convhull)

plot!(phull, c=:red, lab="Convex hull", lw=1.5, alpha=.3)
```

## Convex hull of subsets

```julia
plot(xdom, yval, seriestype=:path, marker=:dot, lab="Data", legend=:bottomright)

partition = [1:20, 21:40, 41:60, 61:80]
convhull = [convex_hull(points[s]) for s in partition]
phull = VPolygon.(convhull)

plot!(phull, c=:red, lab="", lw=1.5, alpha=.3)
```
